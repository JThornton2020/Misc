```python
def checkCode(code):
    #I looked up the value in the system that the guide file said the data was using for the OD code
    OD = 242253008
    #Catch NaN values and throw them out
    if pandas.isna(code):
        return False
    #Cast as an int to compare to OD code
    code= int(code)
    if code == OD:
        return True
    else:
        return False
```

```python
def encountersCheck(date):
    startingSplit = date.split(" ")
    Year, Month, Day = startingSplit[0].split("-")

    Year = int(Year)
    Month = int(Month)
    Day = int(Day)

    if Year < 1999:
        return False
    elif Month < 7:
        return False
    elif Day <=15:
        return False
    else:
        return True
```

```python
def numOpioids(Id, enc):
    med = 0
    opiods = ['Hydromorphone','Fentanyl', 'Oxycodone-acetaminophen']
    listOfMeds = medications.loc[medications["PATIENT"] == Id]
    for thing in listOfMeds:
        if opiods[0] in thing['DESCRIPTION']:
            if activeMed(enc,thing['START'],thing['STOP']):
                med+=1
        if opiods[1] in thing['DESCRIPTION']:
            if activeMed(enc,thing['START'],thing['STOP']):
                med+=1
        if opiods[2] in thing['DESCRIPTION']:
            if activeMed(enc,thing['START'],thing['STOP']):
                med+=1
    return med
```

```python
def readmission(Id, refDate):
    admissions = encounters.loc[encounters['PATIENT'] == Id]
    i=0
    daysApart = 99999
    while i < len(admissions):
        prev = admissions['STOP'][i]
        duration = refDate - prev
        if duration > 0 :
            if daysApart > (duration.days):
                daysApart = duration.days
        i+=1
    return daysApart
```

```python
def opiateRead(Id):
    admissions = encounters.loc[encounters['PATIENT'] == Id]
    i=0
    while i < len(admissions)-1:
        if checkCode(admissions['CODE'][i]):
            checkOut = admissions['STOP'][i]
            if checkCode(admissions['CODE'][i+1]):
                checkIn = admissions['START'][i+1]
                delta = checkIn - checkOut
                deltaDays = delta.days
                if (deltaDays) <= 90:
                    return checkIn
    return 'N/A'
```
